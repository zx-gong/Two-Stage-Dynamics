# 数据分布/矩阵

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成一个随机矩阵
np.random.seed(0)
size = 10
A = np.random.rand(size, size)

# 计算特征值和特征向量
eigenvalues, eigenvectors = np.linalg.eig(A)

# 保留全部特征值的矩阵
full_matrix = np.dot(eigenvectors, np.diag(eigenvalues)).dot(np.linalg.inv(eigenvectors))

# 仅保留1%较小特征值的矩阵
threshold_index = int(0.01 * size)
sorted_indices = np.argsort(eigenvalues)
smallest_indices = sorted_indices[:threshold_index]
smallest_eigenvalues = np.zeros(size)
smallest_eigenvalues[smallest_indices] = eigenvalues[smallest_indices]
small_matrix = np.dot(eigenvectors, np.diag(smallest_eigenvalues)).dot(np.linalg.inv(eigenvectors))

# 提取实部
full_matrix_real = np.real(full_matrix)
small_matrix_real = np.real(small_matrix)

# 绘制热图
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# 热图：保留全部特征值的矩阵
cax1 = ax[0].imshow(full_matrix_real, cmap='Blues', interpolation='nearest')
ax[0].set_title('Full Matrix (Real Part)')
plt.colorbar(cax1, ax=ax[0])

# 热图：仅保留1%较小特征值的矩阵
cax2 = ax[1].imshow(small_matrix_real, cmap='Blues', interpolation='nearest')
ax[1].set_title('Matrix with 1% Smallest Eigenvalues (Real Part)')
plt.colorbar(cax2, ax=ax[1])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成一个随机矩阵
np.random.seed(0)
size = 50
random_matrix = np.random.rand(size, size)
symmetric_matrix = np.dot(random_matrix, random_matrix.T)

# 计算特征值和特征向量
eigenvalues, eigenvectors = np.linalg.eig(symmetric_matrix)

# 设置不同的保留比例
percentages = [0.01, 1, 10, 20, 80, 100]
small_matrices = []

for p in percentages:
    threshold_index = max(1, int(p / 100 * size))  # 确保至少保留一个特征值
    sorted_indices = np.argsort(eigenvalues)
    smallest_indices = sorted_indices[:threshold_index]
    
    smallest_eigenvalues = np.zeros(size)
    smallest_eigenvalues[smallest_indices] = eigenvalues[smallest_indices]
    small_matrix = np.dot(eigenvectors, np.diag(smallest_eigenvalues)).dot(np.linalg.inv(eigenvectors))
    
    # 提取实部
    small_matrices.append(np.real(small_matrix))

# 绘制热图
fig, ax = plt.subplots(1, len(percentages), figsize=(15, 5))

for i, (matrix, p) in enumerate(zip(small_matrices, percentages)):
    cax = ax[i].imshow(matrix, cmap='Blues', interpolation='nearest')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
np.random.seed(0)
w_star = np.array([1, 1])  # 设置 w^\star 为倾斜向量 (45度)
gamma_0 = 1
n_samples = 200

# 采样函数
def sample_data(w_star, gamma_0, n_samples, label):
    z = np.random.randn(n_samples, 2)
    if label == 1:
        z = z[np.dot(z, w_star) >= 0]  # 满足 <w^\star, z> >= 0
        x = gamma_0 * w_star + z
    else:
        z = z[np.dot(z, w_star) <= 0]  # 满足 <w^\star, z> <= 0
        x = -gamma_0 * w_star + z
    return x

# 生成数据
x_P1 = sample_data(w_star, gamma_0, n_samples, label=1)
x_P_minus_1 = sample_data(w_star, gamma_0, n_samples, label=-1)

# 作图
plt.figure(figsize=(6, 7))
plt.scatter(x_P1[:, 0], x_P1[:, 1], color='#CC0000', label='$P_1$', alpha=0.6)
plt.scatter(x_P_minus_1[:, 0], x_P_minus_1[:, 1], color='#0066CC', label='$P_{-1}$', alpha=0.7)

# 标注 w^\star 的方向
plt.quiver(0, 0, w_star[0], w_star[1], angles='xy', scale_units='xy', scale=1, color='green', label='$w^\\star$', width=0.005)

# 分类边界（w^\star 的垂直方向）
slope = -w_star[0] / w_star[1]  # 垂直于 w^\star 的线的斜率
intercept = 0  # 假设边界通过原点
x_vals = np.linspace(-3, 3, 100)
boundary_line = slope * x_vals + intercept

plt.plot(x_vals, boundary_line, 'g:', label='Decision Boundary')

# 绘制分类间隔线（距离分类边界的平行线）
margin_offset = gamma_0 / np.linalg.norm(w_star)  # 分类间隔
boundary_line_margin1 = slope * x_vals + intercept + margin_offset
boundary_line_margin2 = slope * x_vals + intercept - margin_offset

plt.plot(x_vals, boundary_line_margin1, ':', color='black', label='Margin (+$\gamma_0$)')
plt.plot(x_vals, boundary_line_margin2, ':', color='black', label='Margin (-$\gamma_0$)')

# plt.xlim(-3, 3)
# plt.ylim(-3, 3)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('classification_visualization_2.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sample_data(w_star, gamma_0, n_samples, label):
    z = np.random.randn(n_samples, 2)
    if label == 1:
        z = z[np.dot(z, w_star) >= 0]  # 满足 <w^\star, z> >= 0
        x = gamma_0 * w_star + z
    else:
        z = z[np.dot(z, w_star) <= 0]  # 满足 <w^\star, z> <= 0
        x = -gamma_0 * w_star + z
    return x

# 生成第一部分数据
w_star = np.array([1, 0])  # 假设 w_star 为单位向量
gamma_0 = 1
n_samples = 100

# 正类和负类
data_pos = sample_data(w_star, gamma_0, n_samples, label=1)
data_neg = sample_data(w_star, gamma_0, n_samples, label=-1)

# 第二部分数据
z = np.random.randn(n_samples, 2)
z_pos = z  # z作为正类
z_neg_1 = z - 0.5  # z - zeta 作为负类
z_neg_2 = z + 0.5  # z + zeta 作为负类

# 将数据合并
data_full = np.concatenate([data_pos, data_neg, z_pos, z_neg_1, z_neg_2], axis=0)

# 创建标签用于可视化（1：正类，-1：负类）
labels_full = np.concatenate([np.ones(len(data_pos)), -np.ones(len(data_neg)), 
                              np.ones(len(z_pos)), -np.ones(len(z_neg_1)), -np.ones(len(z_neg_2))])

# 可视化
plt.figure(figsize=(6, 6))
plt.scatter(data_full[labels_full == 1, 0], data_full[labels_full == 1, 1], color='#CC0000', alpha=0.6, label='Positive Class')
plt.scatter(data_full[labels_full == -1, 0], data_full[labels_full == -1, 1], color='#0066CC', alpha=0.7, label='Negative Class')
# 隐藏坐标轴
plt.axis('off')

# 添加图例
plt.tight_layout()

# 保存为高清 PNG 图片
plt.savefig('combined_classification_visualization.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
